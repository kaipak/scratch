```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def weird_func(x):
    x[x == 0] = -1
    y = np.cumsum(x)
    return y

arr_size = 100000
x = np.arange(0, arr_size)
y_pre = np.random.randint(0, 2, arr_size)
y = weird_func(y_pre)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 9))

plt.scatter(x, y)
```
